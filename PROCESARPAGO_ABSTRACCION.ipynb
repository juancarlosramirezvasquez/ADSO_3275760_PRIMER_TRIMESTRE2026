```python
# ALGORITMO METODO DE PAGO
from abc import ABC, abstractmethod


# CLASE PARA TODAS LAS CLASES HIJAS
class GenerarReporte(ABC):

    @abstractmethod
    def generar_reporte(self):
        pass


# CLASE BASE DEL SISTEMA DE PAGO
class SistemaPago(ABC):

    def __init__(self, monto: float) -> None:
        self.monto = monto

    @abstractmethod
    def procesar_pago(self) -> None:
        pass


class PagoBancario(SistemaPago, GenerarReporte):

    def procesar_pago(self) -> None:
        print(f"Procesando pago bancario de ${self.monto:,.2f}")

    def generar_reporte(self) -> None:
        print(f"Generando reporte de pago bancario de ${self.monto:,.2f}")


class PagoCriptomoneda(SistemaPago):

    def procesar_pago(self) -> None:
        print(f"Procesando pago de criptomoneda de ${self.monto:,.2f}")


# FUNCION POLIMORFICA
def ejecutar_pago(pago: SistemaPago) -> None:
    pago.procesar_pago()


def main():
    pago1 = PagoBancario(1000)
    pago2 = PagoCriptomoneda(2000)

    ejecutar_pago(pago1)
    ejecutar_pago(pago2)

    pago1.generar_reporte()


if __name__ == "__main__":
    main()
```
