```python
#ejercicio zoologico
#vamos a construir un modelo zoologico donde existen animales genericos (clase abstacta)
# y tendremos animales especificos que hereden comportamientos e implementala accion hacer el sonido mediante poliforfismo

from abc import ABC, abstractmethod
from typing import List

class Movible(ABC):
    @abstractmethod
    def mover(self):
        pass

class Animal(ABC):
  def __init__(self, nombre: str) -> None:
    #esta el atributo privado y encapsulado
    self.__nombre = "" # Initialize the private attribute
    self.nombre = nombre # This will call the setter

  @property #decorador  , hacer getter ,  muestra  el nombre de forma segura
  def nombre(self) -> str:
    return self.__nombre

  @nombre.setter
  def nombre(self, valor: str) -> None:#aseguramos que el nombre venga de una cadena de texto valida
    if isinstance(valor, str) and valor.strip(): # Corrected isinstance check
      self.__nombre = valor.strip().title() #asigna el nuevo nombre quita espacio en blanco y poner primera letra mayuscula
    else:
      raise ValueError("El nombre debe ser una cadena de texto valida")

  @abstractmethod
  def sonido(self) -> None:
    pass


#CREAREMOS LAS CLASES HIJAS
class Perro(Animal):
  def sonido(self) -> None:
    print(f"\n{self.nombre} (perro) dice:  Guaguaguaooooo")

class Gato(Animal):
  def sonido(self) -> None:
    print(f"{self.nombre} (gato) dice: Miaaaaauuuuuuuu ")

class Vaca(Animal):
  def sonido(self) -> None:
    print(f"{self.nombre} (vaca) dice:  MUUUUUUUUUUU ")

class Oveja(Animal):
  def sonido(self) -> None:
    print(f"{self.nombre} (obeja) dice: Meeeeeeggg")

#**********************************
class Leon(Animal, Movible): # Leon now inherits from Movible
  def sonido(self) -> None:
    print(f"{self.nombre} (leon) dice: Roooooaarrrr")

  def mover(self) -> None:
    print(f"{self.nombre} CAMINA SIGILOSAMENTE POR LA SABANA...")


#FUNCIONES POLIMORFICAS
def hacer_sonido(animal: Animal) -> None:
  animal.sonido()

def desplazar(animal: Movible) -> None: # Typo Movile corrected to Movible
  animal.mover()

def main() -> None:
  try:
    animales: List[Animal] = [
        Perro("Rocky"),
        Gato("Misu"),
        Vaca("Lola"),
        Oveja("Blanca"),
        Leon("Simba"),
    ]

    print("POLIMORFISMO EN MODELO DE ZOOLOGICO\n")
    for animal in animales:
      hacer_sonido(animal)
    print("\nMOVIMIENTO")
    for animal in animales:
      if isinstance(animal, Movible): # Typo Movile corrected to Movible
        desplazar(animal)


        #CAMBIO DE NOMBRE

    print("\nACTUALIZACION DE LOS NOMBRES")
    animales[0].nombre = "Luna"
    animales[1].nombre = "Firulais"
    print(f"\nEl nuevo nombre del perro es: {animales[0].nombre}")
    print(f"El nuevo nombre del gato es: {animales[1].nombre}")
    for animal in animales:
      hacer_sonido(animal)





  except ValueError as e:
    print(f"Error de validación: {e}")


if __name__ == "__main__":
  main()
```
