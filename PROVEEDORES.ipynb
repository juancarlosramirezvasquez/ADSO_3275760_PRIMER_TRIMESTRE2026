```python
class Proveedor:
    def __init__(self,nombre, producto, precio, stock, ubicacion):
        self.__nombre = nombre
        self.__producto = producto
        self.__precio = precio
        self.__stock = stock
        self.__ubicacion = ubicacion



    def get_nombre(self):
        return self.__nombre

    def get_producto(self):
        return self.__producto

    def get_precio(self):
        return self.__precio

    def get_stock(self):
        return self.__stock


    def get_ubicacion(self):
        return self.__ubicacion

    def set_nombre(self, nuevo_nombre):
        if isinstance(nuevo_nombre, str) and nuevo_nombre != "" and nuevo_nombre.strip() :
            self.__nombre = nuevo_nombre.title()
        else:
            raise ValueError("El nombre debe ser un texto válido no vacío.")

    def set_producto(self, nuevo_producto):
        if isinstance(nuevo_producto, str) and nuevo_producto != "" and nuevo_producto.strip():
            self.__producto = nuevo_producto
        else:
            raise ValueError("El producto debe ser un texto válido no vacío.")



    def set_ubicacion(self, nueva_ubicacion):
        if isinstance(nueva_ubicacion, (int, float)) and nueva_ubicacion >= 0:
            self.__ubicacion = nueva_ubicacion
        else:
            raise ValueError("La ubicación debe ser un número positivo.")

    def mostrar_informacion(self):
        print(f"nombre: {self.__nombre}")
        print(f"Producto: {self.__producto}")
        print(f"Precio: {self.__precio:,.2f}")
        print(f"stock: {self.__stock}")
        print(f"ubicacion: {self.__ubicacion}")

def main():
    print("\nSISTEMA DE PROVEEDORES\n")
    proveedor1 = Proveedor("ANDRES", "CAMISAS", 32.500 , 15 , 356899)
    proveedor2 = Proveedor("SANDRA", "PANTALONES", 100.000, 12 , 246638)
    proveedor3 = Proveedor("CAMILA", "CHAQUETAS", 25.999 , 18 , 51242638)

    print("\nLOS DATOS DE LOS PROVEEDORES SON")
    print("\nLOS DATOS DEL PROVEEDOR 1 SON")
    proveedor1.mostrar_informacion()
    print("\nLOS DATOS DEL PROVEEDOR 2 SON")
    proveedor2.mostrar_informacion()
    print("\nLOS DATOS DEL PROVEEDOR 3 SON")
    proveedor3.mostrar_informacion()


    print("\nACTUALIZANDO LOS DATOS DE LOS PROVEEDORES\n")
    proveedor1.set_nombre("andres gonzales")
    proveedor2.set_nombre("SANDRA RAMIREZ")
    proveedor3.set_nombre("CAMILA TORRES")

    proveedor1.set_producto("CAMISAS DEPORTIVAS")
    proveedor2.set_producto("PANTALONES DE MEZCLILLA")
    proveedor3.set_producto("CHAQUETAS DE INVIERNO")

    proveedor1.set_ubicacion(3738839)
    proveedor2.set_ubicacion(6388390)
    proveedor3.set_ubicacion(6383990)

    print("\nLOS DATOS ACTUALIZADOS DE LOS PROVEEDORES SON ")
    print("\nLOS DATOS ACTUALIZADOS DEL PROVEEDOR 1 SON ")
    proveedor1.mostrar_informacion()
    print("\nLOS DATOS ACTUALIZADOS DEL PROVEEDOR 2 SON ")
    proveedor2.mostrar_informacion()
    print("\nLOS DATOS ACTUALIZADOS DEL PROVEEDOR 3 SON ")
    proveedor3.mostrar_informacion()




if __name__ == "__main__":
    main()
```
