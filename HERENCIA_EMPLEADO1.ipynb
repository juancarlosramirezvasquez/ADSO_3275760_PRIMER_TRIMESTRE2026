```python
class Empleado:
    def __init__(self, nombre, salario):
        self.nombre = nombre
        self.salario = salario
#creamos metodo a trabajar
    def trabajar(self):
        print(f"{self.nombre} esta realizando un trabajo de la empresa")

class Gerente(Empleado):
    def __init__(self, nombre, salario, departamento):
        super().__init__(nombre, salario)
        self.departamento = departamento
    def mostrar_info(self):
        print(f"NOMBRE: El atributo del padre: {self.nombre}")
        print(f"SALARIO: El atributo del padre: {self.salario}")
        print(f"DEPARTAMENTO: El atributo del hijo: {self.departamento}")
class vendedor(Empleado):
    def __init__(self, nombre, salario, ciudad, comision):
        super().__init__(nombre, salario)
        self.ciudad = ciudad
        self.comision = comision
    def mostrar_info(self):
        print(f"\nNOMBRE: {self.nombre}")
        print(f"SALARIO: {self.salario}")
        print(f"CIUDAD: {self.ciudad}")
        print(f"COMISION: {self.comision}")
def main():
    Gerente1= Gerente("carlos", 12334312323, "Kennedy")
    Gerente1.mostrar_info()
    vendedor1 = vendedor("eduardo", 2000000, "Bogota", "comicion del 20 porciento")
    vendedor1.mostrar_info()

if __name__ == "__main__":
    main()
```
