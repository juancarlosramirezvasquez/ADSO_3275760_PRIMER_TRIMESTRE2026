```python
class empleado:
    def __init__(self, nombre, salario_base):
        self.nombre = nombre
        self.salario_base = salario_base

    def validar(self):
        if not isinstance(self.salario_base, (int, float)) or self.salario_base < 0:
            raise ValueError("\nEl salario debe ser un valor numerico y positivo")

    def mostrar_datos(self):
        print(f"\nEmpleado: {self.nombre}")
        print(f"\nSalario base: {self.salario_base}")

    def calcular_salario(self):
        print("\n==Calculando el salario del empleado....==")


class empleadotiempocompleto(empleado):
    def __init__(self, nombre, salario_base, bono):
        super().__init__(nombre, salario_base)
        self.bono = bono

    def calcular_salario(self):
        super().validar()
        super().mostrar_datos()
        super().calcular_salario()
        salario_final = self.salario_base + self.bono
        print(f"\nSalario final con bono: {salario_final}")


class empleadoporhoras(empleado):
    def __init__(self, nombre, horas_trabajadas, valor_hora):
        super().__init__(nombre, 0)
        self.horas_trabajadas = horas_trabajadas
        self.valor_hora = valor_hora

    def calcular_salario(self):
        super().calcular_salario()
        salario_final = self.horas_trabajadas * self.valor_hora
        print(f"Salario por horas: {salario_final}")


def main():
    nombre = input("\nIngrese el nombre: ")
    print("\n1. Tiempo completo")
    print("2. Por horas")
    opcion = int(input("\nSeleccione tipo de empleado: "))

    if opcion == 1:
        salario = float(input("\nIngrese el salario base: "))
        bono = float(input("Ingrese el bono: "))
        empleado = empleadotiempocompleto(nombre, salario, bono)
    else:
        horas = float(input("\nIngrese horas trabajadas: "))
        valor = float(input("\nIngrese el valor por horas: "))
        empleado = empleadoporhoras(nombre, horas, valor)

    empleado.calcular_salario()


if __name__ == "__main__":
    main()

```
