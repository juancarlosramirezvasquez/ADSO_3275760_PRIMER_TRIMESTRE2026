```python
class  libro: #El nombre de la clase
    def __init__(self, titulo, precio): #El contructor
        self.__titulo = titulo
        self.__precio = precio   # los atributos

    @property  #Es El decorador  ,con este comando creo el get  y muestra la informacion de loos atributos
    def precio(self):
        return self.__precio

    @property
    def titulo(self):
        return self.__titulo

    @precio.setter  #El decorador  set para modificar el valor del atributo
    def precio(self, nuevo_precio):
        if isinstance(nuevo_precio, (int, float)) and nuevo_precio >= 0:
         self.__precio = nuevo_precio
        else:
         print("DEBE INGRESAR UN NUMERO VALIDO EN EL PRECIO")

    @titulo.setter
    def titulo(self, nuevo_titulo):
        if isinstance(nuevo_titulo, str):
         self.__titulo = nuevo_titulo
        else:
         print("DEBE INGRESAR UN TITULO VALIDO")

    def mostrar_informacion(self): #metodo para mostrar la informacion del libro
        print(f"\nEl titulo del libro es: {self.__titulo}")
        print(f"El precio del libro es: {self.__precio}\n")


def main():
    print("==== CREACION DE OBJETOS EN POO ====")
    libro1 = libro("EL SEÑOR DE LOS ANILLOS", 150.000)#Creo mi primer objeto
    libro2 = libro("EL HOBBIT", 120.000) #Creo mi segundo objeto

    print("EL NOMBRE DEL PRIMER LIBRO ES:",libro1.titulo) #Muestro el titulo del primer libro
    print("EL PRECIO DEL PRIMER LIBRO ES:", libro1.precio) #Muestro el precio del primer libr

    print("EL NOMBRE DEL SEGUNDO LIBRO ES:",libro2.titulo) #Muestro el titulo del primer libro
    print("EL PRECIO DEL SEGUNDO LIBRO ES:", libro2.precio) #Muestro el precio del segundo libr


    print("TODOS LOS DATOS DEL PRIMER LIBRO SON:")
    libro1.mostrar_informacion()

    print("TODOS LOS DATOS")
    libro2.mostrar_informacion()

    print("\nACTUALIZAREMOS EL TITULO DEL PRIMER LIBRO")
    libro1.titulo = input("Ingrese el nuevo titulo del libro:") #Actualizo el titulo del libro1
    print("TITULO ACTUALIZADO: Ahora  el primer libro se llama:", libro1.titulo)

    print("\nACTUALIZAREMOS EL TITULO DEL SEGUNDO LIBRO")
    libro2.titulo = input("Ingrese el nuevo titulo del segundo libro:") #Actualizo el titulo del libro2
    print("TITULO ACTUALIZADO: Ahora  el segundo libro se llama:", libro2.titulo)

    print("\nACTUALIZAREMOS EL PRECIO DEL PRIMER LIBRO")
    libro1.precio = float(input("Ingrese el nuevo precio del libro:")) #Actualizo el precio del libro1
    print("PRECIO ACTUALIZADO: Ahora  el precio del libro es:", libro1.precio)

    print("\nACTUALIZAREMOS EL PRECIO DEL SEGUNDO LIBRO")
    libro2.precio = float(input("Ingrese el nuevo precio del segundo libro:")) #Actualizo el precio del libro1
    print("PRECIO ACTUALIZADO: Ahora  el precio del libro segundo es:", libro2.precio)

    print("\nLOS NUEVOS DATOS DEL PRIMER  LIBRO SON:")
    libro1.mostrar_informacion()

    print("\nLOS NUEVOS DATOS DEL SEGUNDO LIBRO SON:")
    libro2.mostrar_informacion()

if __name__ == "__main__":
   main()
```
