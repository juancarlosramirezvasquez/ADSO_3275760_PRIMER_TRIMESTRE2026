```python
class Padre:
    def __init__(self, mensaje, color_ojos, tipo_sangre):
        self.mensaje = mensaje
        self.color_ojos = color_ojos
        self.tipo_sangre = tipo_sangre



class Hijo(Padre):
    def __init__(self, mensaje,color_ojos,tipo_sangre,nombre, genero, estatura):
        super().__init__(mensaje,color_ojos,tipo_sangre) # REUTILIZA EL CONSTRUCTOR DE LA CLASE PADRE para 'mensaje'
        self.nombre = nombre
        self.genero = genero
        self.estatura = estatura



def main():
    obj = Hijo("Hola desde la clase padre", "Verdes", "O+", "Juan", "Masculino", 1.75) #instanciamos un objeto de la clase hijo
    print(f"{obj.mensaje}, {obj.nombre}, {obj.color_ojos}, {obj.tipo_sangre}, {obj.genero}, {obj.estatura}") #Accedemos al atributo mensaje heredado de la clase padre
    print(obj.mensaje) #Accedemos al atributo mensaje heredado de la clase padre
    print(obj.nombre)


if __name__ == "__main__":
    main()
```
