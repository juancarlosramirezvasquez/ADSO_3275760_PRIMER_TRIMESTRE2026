```python
from abc import ABC, abstractmethod

# CLASE ABSTRACTA
class Figura(ABC):

    @abstractmethod
    def calcular_area(self):
        pass


# CLASE CIRCULO
class Circulo(Figura):

    def __init__(self, radio):
        self.radio = radio

    def calcular_area(self):
        return 3.14159 * self.radio * self.radio


# CLASE CUADRADO
class Cuadrado(Figura):

    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        return self.base * self.altura


# FUNCION POLIMORFICA
def mostrar_area(figura: Figura):
    print(f"El área de la figura es: {figura.calcular_area():,.2f}")


def main():
    circulo = Circulo(5)  # creación del objeto
    mostrar_area(circulo)  # polimorfismo

    cuadrado = Cuadrado(4, 5)
    mostrar_area(cuadrado)


if __name__ == "__main__":
    main()
```
