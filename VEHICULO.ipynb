```python
class Vehiculo: #Clase padre
    def mover(self):
        print("\nAqui muestro el metodo de la clase padre")

class Carro(Vehiculo): #Clase hija que hereda el metodo  de vehiculo
    pass

class Moto(Vehiculo): #Clase hija que hereda el metodo  de vehiculo
    pass

def main():

  vehiculo1 = Vehiculo() #Creo un objeto de la clase padre
  vehiculo1.mover() #Llamo al metodo mover del objeto de la clase padre

  print("La clase hijo repetira carro el metodo mover() de la clase padre")
  carro1 = Carro() #Creo un objeto de la clase hija
  carro1.mover() #Llamo al metodo mover del objeto de la clase hija

  print("La clase hijo repetira moto el metodo  mover() de la clase padre")
  moto1 = Moto() #Creo un objeto de la clase hija
  moto1.mover() #Llamo al metodo mover del objeto de la clase hija



if __name__ == "__main__":
    main()
```
