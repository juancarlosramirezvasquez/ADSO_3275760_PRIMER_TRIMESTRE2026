```python
class  Mueble: #El nombre de la clase
    def __init__(self, material, color, precio, forma, estilo): #El contructor
        self.__material = material
        self.__color = color   # los atributos
        self.__precio = precio
        self.__forma = forma
        self.__estilo = estilo

    @property  #Es El decorador  ,con este comando creo el get  y muestra la informacion de loos atributos
    def material(self):
        return self.__material

    @property
    def color(self):
        return self.__color

    @property
    def precio(self):
        return self.__precio

    @property
    def forma(self):
        return self.__forma

    @property
    def estilo(self):
        return self.__estilo



    @material.setter  #El decorador  set para modificar el valor del atributo
    def material(self, nuevo_material):
        if isinstance(nuevo_material, str):
         self.__material = nuevo_material
        else:
         print("DEBE INGRESAR UN MATERIAL VALIDO")

    @color.setter  #El decorador  set para modificar el valor del atributo
    def color(self, nuevo_color):
        if isinstance(nuevo_color, str):
         self.__color = nuevo_color
        else:
         print("DEBE INGRESAR UN COLOR VALIDO")

    @precio.setter  #El decorador  set para modificar el valor del atributo
    def precio(self, nuevo_precio):
        if isinstance(nuevo_precio, (int, float)) and nuevo_precio >= 0:
         self.__precio = nuevo_precio
        else:
         print("DEBE INGRESAR UN PRECIO VALIDO")

    @forma.setter  #El decorador  set para modificar el valor del atributo
    def forma(self, nueva_forma):
        if isinstance(nueva_forma, str):
         self.__forma = nueva_forma
        else:
         print("DEBE INGRESAR UNA FORMA VALIDA")

    @estilo.setter  #El decorador  set para modificar el valor del atributo
    def estilo(self, nuevo_estilo):
        if isinstance(nuevo_estilo, str):
         self.__estilo = nuevo_estilo
        else:
         print("DEBE INGRESAR UN ESTILO VALIDO")

    def mostrar_informacion(self): #metodo para mostrar la informacion del mueble
        print(f"\nEl material del mueble es: {self.__material}")
        print(f"El color del mueble es: {self.__color}")
        print(f"El precio del mueble es: {self.__precio}")
        print(f"El estilo del mueble es: {self.__estilo}")
        print(f"La forma del mueble es: {self.__forma}\n")


def main():
    print("\n==== CREACION DE MUEBLES EN POO ====\n")
    mueble1 = Mueble("Madera", "Castaño", 150.000, "Rectangular", "Clásico")#Creo mi primer objeto
    mueble2 = Mueble("Metal", "Plateado", 120.000, "Circular", "Moderno") #Creo mi segundo objeto

    print("EL MATERIAL DEL PRIMER MUEBLE ES:",mueble1.material) #Muestro el material del primer mueble
    print("EL COLOR DEL PRIMER MUEBLE ES:", mueble1.color) #Muestro el color del primer mueble
    print("EL PRECIO DEL PRIMER MUEBLE ES:", mueble1.precio) #Muestro el precio del primer mueble
    print("LA FORMA DEL PRIMER MUEBLE ES:", mueble1.forma) #Muestro la forma del primer mueble
    print("EL ESTILO DEL PRIMER MUEBLE ES:", mueble1.estilo) #Muestro el estilo del primer mueble

    print("\nEL MATERIAL DEL SEGUNDO MUEBLE ES:",mueble2.material) #Muestro el material del segundo mueble
    print("EL COLOR DEL SEGUNDO MUEBLE ES:", mueble2.color) #Muestro el color del segundo mueble
    print("EL PRECIO DEL SEGUNDO MUEBLE ES:", mueble2.precio) #Muestro el precio del segundo mueble
    print("LA FORMA DEL SEGUNDO MUEBLE ES:", mueble2.forma) #Muestro la forma del segundo mueble
    print("EL ESTILO DEL SEGUNDO MUEBLE ES:", mueble2.estilo) #Muestro el estilo del segundo mueble

    print("\nTODOS LOS DATOS DEL PRIMER MUEBLE SON\n")
    mueble1.mostrar_informacion()

    print("\nTODOS LOS DATOS DEL SEGUNDO MUEBLE SON\n")
    mueble2.mostrar_informacion()

    print("\nACTUALIZAREMOS EL COLOR DEL PRIMER MUEBLE")
    mueble1.color = input("Ingrese el nuevo color del mueble:") #Actualizo el color del mueble1
    print("COLOR ACTUALIZADO: Ahora  el primer mueble es de color:", mueble1.color)

    print("\nACTUALIZAREMOS EL COLOR DEL SEGUNDO MUEBLE")
    mueble2.color = input("Ingrese el nuevo color del segundo mueble:") #Actualizo el color del mueble2
    print("COLOR ACTUALIZADO: Ahora  el segundo mueble es de color:", mueble2.color)

    print("\nACTUALIZAREMOS EL PRECIO DEL PRIMER MUEBLE")
    mueble1.precio = float(input("Ingrese el nuevo precio del mueble:")) #Actualizo el precio del mueble1
    print("PRECIO ACTUALIZADO: Ahora  el precio del mueble es:", mueble1.precio)

    print("\nACTUALIZAREMOS EL PRECIO DEL SEGUNDO MUEBLE")
    mueble2.precio = float(input("Ingrese el nuevo precio del segundo mueble:")) #Actualizo el precio del mueble2
    print("PRECIO ACTUALIZADO: Ahora  el precio del segundo mueble es:", mueble2.precio)

    print("\nACTUALIZAREMOS EL ESTILO DEL PRIMER MUEBLE")
    mueble1.estilo = input("Ingrese el nuevo estilo del mueble:") #Actualizo el estilo del mueble1
    print("ESTILO ACTUALIZADO: Ahora  el primer mueble es de estilo:", mueble1.estilo)

    print("\nACTUALIZAREMOS EL ESTILO DEL SEGUNDO MUEBLE")
    mueble2.estilo = input("Ingrese el nuevo estilo del segundo mueble:") #Actualizo el estilo del mueble2
    print("ESTILO ACTUALIZADO: Ahora  el segundo mueble es de estilo:", mueble2.estilo)


    print("\nLOS NUEVOS DATOS DEL PRIMER  MUEBLE SON:")
    mueble1.mostrar_informacion()

    print("\nLOS NUEVOS DATOS DEL SEGUNDO MUEBLE SON:")
    mueble2.mostrar_informacion()

if __name__ == "__main__":
   main()
```
