```python
class Empleado:
    def __init__(self, nombre, salario):
        self.nombre = nombre
        self.salario = salario

class Gerente(Empleado):
    def __init__(self, nombre, salario, departamento):
        super().__init__(nombre, salario)  # Reutiliza el constructor de la clase Empleado
        self.departamento = departamento

    def mostrar_info(self):
      print(f"\nNombre: {self.nombre}\nSalario: ${self.salario:,.0f}, \nDepartamento: {self.departamento}")

def main():
    gerente1 = Gerente("Ana", 75000, "Innovación")
    gerente1.mostrar_info()

if __name__ == "__main__":
    main()
```
