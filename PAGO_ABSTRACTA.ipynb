```python
from abc import ABC, abstractmethod

# Clase abstracta para generar reportes
class GenerarReporte(ABC):
    @abstractmethod
    def generar_reporte(self) -> None:
        pass


# Clase abstracta del sistema de pago
class SistemaPago(ABC):
    def __init__(self, monto: float) -> None:
        self.monto = monto

    @abstractmethod
    def procesar_pago(self) -> None:
        pass


# Clase hija para pago bancario
class PagoBancario(SistemaPago, GenerarReporte):

    def procesar_pago(self) -> None:
        print(f"\nProcesado transferencia bancaria por ${self.monto:,.2f}")

    def generar_reporte(self) -> None:
        print(f"Reporte de transferencia bancaria por ${self.monto:,.2f}")


# Clase hija para pago con cripto
class PagoCripto(SistemaPago):

    def procesar_pago(self) -> None:
        print(f"Procesado pago con cripto por ${self.monto:,.2f}")


# Función que ejecuta el pago
def ejecutar_pago(pago: SistemaPago) -> None:
    pago.procesar_pago()


# Programa principal
def main() -> None:
    print("************ Sistema de Pagos ************")

    pago1 = PagoBancario(1000)
    ejecutar_pago(pago1)
    pago1.generar_reporte()

    pago2 = PagoCripto(500)
    ejecutar_pago(pago2)


if __name__ == "__main__":
    main()
```
