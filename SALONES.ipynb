```python
class Salon:
    def __init__(self, Codigo, Nombre, Responsable,Puestos):
        self.Codigo = Codigo
        self.Nombre = Nombre
        self.Responsable = Responsable
        self.Puestos = Puestos
Salon1 = Salon("23456", "301", "JUAN RAMIREZ", "24 PUESTOS ")
Salon2 = Salon("25637", "304", "KAREN RODRIGUEZ", "32 PUESTOS ")
Salon3 = Salon("13897", "331", "LUIS PEREZ", "16 PUESTOS ")
Salon4 = Salon("67748", "231", "ALEX HERNANDEZ", "25 PUESTOS ")

print(f"\nRESPONSABLE:{Salon1.Responsable} \nNOMBRE:{Salon1.Nombre} \nPUESTOS:{Salon1.Puestos} \nCODIGO:{Salon1.Codigo} ")
print(f"\nRESPONSABLE:{Salon2.Responsable} \nNOMBRE:{Salon2.Nombre} \nPUESTOS:{Salon2.Puestos} \nCODIGO:{Salon2.Codigo} ")
print(f"\nRESPONSABLE:{Salon3.Responsable} \nNOMBRE:{Salon3.Nombre} \nPUESTOS:{Salon3.Puestos} \nCODIGO:{Salon3.Codigo} ")
print(f"\nRESPONSABLE:{Salon4.Responsable} \nNOMBRE:{Salon4.Nombre} \nPUESTOS:{Salon4.Puestos} \nCODIGO:{Salon4.Codigo} ")
```
